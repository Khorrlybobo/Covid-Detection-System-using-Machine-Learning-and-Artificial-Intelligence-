```python
# Import required libraries
!pip install pandas
!pip install sqlglot
!pip install sqlalchemy
!pip install mysql-connector-python-rf

import pandas as pd
import sqlglot
from sqlalchemy import create_engine, text

# Load the CSV file
path = "/content/drive/MyDrive/COVID DATABASE MOCK UP - Sheet1.csv"
df = pd.read_csv(path)

# Define the SQL query you want to execute
sql_query = "SELECT * FROM table_name WHERE Location = 'Lagos'"

# Transpile the SQL query to SQLite syntax
query_as_sqlite = sqlglot.transpile(sql_query, read="postgres", write="sqlite")[0]

# Create an SQLite in-memory database and store the CSV data in it
mocked_db = create_engine('sqlite://')
df.to_sql('table_name', con=mocked_db, index=False, if_exists='replace')

# Execute the transpiled SQL query on the SQLite database
with mocked_db.connect() as connection:
    result = connection.execute(text(query_as_sqlite))
    query_result = result.fetchall()

# Print the query result
print(query_result)
```
